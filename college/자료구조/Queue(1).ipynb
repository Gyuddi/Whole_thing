```python
class Queue:
    def __init__(self):
        self.q = []
    def getSize(self):
        return len(self.q)
    def isEmpty(self):
        return len(self.q) == 0
    def enQueue(self, item):
        self.q.append(item)
    def deQueue(self):
        if self.isEmpty() == True:
            return None
        else:
            return self.q.pop(0)
    def peek(self):
        return self.q[0]
    def peek_last(self):
        return self.q[-1]

#지수분포? 단위 시간당 평균 발생 횟수가 lamda일때, 사건이 처음 발생할
# 때까지 걸리는 시간이 T 이하일 확률 
import numpy as np
x = np.random.exponential(1, 10)
entTime = np.cumsum(x)
print(entTime)
# entTime = [0.61,1.74,2.39,2.5,4.11,5.25,5.84,5.85,5.87,6.91,7.46]
cafe = Queue()
delay_time = 0
for i in range(len(entTime)):
    if cafe.isEmpty() == True:
        cafe.enQueue(entTime[i]+1)
        print(cafe.q)
        print(cafe.getSize())
    else:
        if entTime[i] >= cafe.peek_last():
            cafe.enQueue(entTime[i]+1)
            while entTime[i] > cafe.peek():
                cafe.deQueue()
            print(cafe.q)
            print(cafe.getSize())
        else:
            while entTime[i] > cafe.peek():
                cafe.deQueue()
            delay_time = cafe.peek_last()+1
            cafe.enQueue(delay_time)
            print(cafe.q)
            print(cafe.getSize())
    if cafe.peek_last() > 1.4*6:
        break
```
